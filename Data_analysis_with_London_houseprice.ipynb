```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/London Housing Data.csv')
data.head()
```

```
#converting datatime column to date time format

data.date= pd.to_datetime(data.date)
```

```
data.info()
```

```
#add a year column to data set 
data['year']= data.date.dt.year
```

```
data.head(2)
```

```
#add month column with month from date to the dataframe
data['month']= data.date.dt.month
```

```
data.head()
```

```
#show the records having no.of crimes is 0
data[data['no_of_crimes'] ==0]
```

```
#what is the max and min avg_price per year in england

data_1 = data[data.area=='england']
```

```
data_1.groupby('year').average_price.agg(['min','max'])
```

```
#no.of min,max crimes per area

data.groupby('area').no_of_crimes.agg(['min','max'])
```
